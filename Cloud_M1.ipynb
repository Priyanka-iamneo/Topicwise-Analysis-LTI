```python
df=pd.read_csv("Cloud(Topic-wise) - All Milestones.csv")
```

```python
fig=px.bar(df,x='Sub-status',color='Topic')
fig.show()
```

```python
milestone1=df.loc[df['Milestone Number']=='Milestone 01']
```

```python
milestone1
```

```python
M1Attempt = milestone1.drop(milestone1.loc[((milestone1['Sub-status']=='Cleared - Attempt 01') | (milestone1['Sub-status']=='Cleared - Attempt 02')) & (milestone1['Marks Scored']==0)].index)
```

```python
M1Attempt
```

```python
M1_topic_questions = M1Attempt.groupby(['Topic', 'SubTopic']).agg({'Total Questions': 'sum',
                                                                               'Total Attended Questions': 'sum',
                                                                               'Questions Correct': 'sum',
                                                                               'Questions Wrong': 'sum',
                                                                               'Questions Partial': 'sum',
                                                                               'Questions Skipped': 'sum',
                                                                               'Questions Not Viewed':'sum'}).reset_index().sort_values(['Total Questions','Questions Correct'],ascending=False)
```

```python
M1_topic_questions['Solved_percentage']=(M1_topic_questions['Questions Correct']/ M1_topic_questions['Total Questions'])*100
```

```python
M1_topic_questions
```

```python
M1Attempt['SubTopic'].nunique()
```

```python
fig = px.bar(M1_topic_questions, y="Topic", x="Total Attended Questions",color='Questions Correct')
fig.show()
```

```python
fig = px.bar(M1_topic_questions, y="Topic", x="Solved_percentage",color='Questions Correct')
fig.show()
```

```python
M1_subtopic_questions = M1Attempt.groupby(['SubTopic']).agg({'Total Questions': 'sum',
                                                                               'Total Attended Questions': 'sum',
                                                                               'Questions Correct': 'sum',
                                                                               'Questions Wrong': 'sum',
                                                                               'Questions Partial': 'sum',
                                                                               'Questions Skipped': 'sum',
                                                                               'Questions Not Viewed':'sum'}).reset_index().sort_values(['Total Questions','Questions Correct'],ascending=False)
```

```python
M1_subtopic_questions['Solved_percentage']=(M1_subtopic_questions['Questions Correct']/ M1_subtopic_questions['Total Questions'])*100
```

```python
fig = px.bar(M1_subtopic_questions, y="SubTopic", x="Total Attended Questions",color='Questions Correct')
fig.show()
```

```python
fig = px.bar(M1_subtopic_questions, y="SubTopic", x="Solved_percentage",color='Questions Correct')
fig.show()
```
